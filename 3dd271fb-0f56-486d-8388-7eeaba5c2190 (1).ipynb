## Описание проекта

Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение, чтобы привлечь новых клиентов. В нём можно будет узнать рыночную стоимость своего автомобиля.

**Задача**: необходимо **построить модель, которая сумеет определять стоимость машины** по признакам. В моём распоряжении данные о технических характеристиках, комплектации и ценах других автомобилей, в их числе следующие столбцы:

**Признаки**:
- DateCrawled — дата скачивания анкеты из базы
- VehicleType — тип автомобильного кузова
- RegistrationYear — год регистрации автомобиля
- Gearbox — тип коробки передач
- Power — мощность (л. с.)
- Model — модель автомобиля
- Kilometer — пробег (км)
- RegistrationMonth — месяц регистрации автомобиля
- FuelType — тип топлива
- Brand — марка автомобиля
- Repaired — была машина в ремонте или нет
- DateCreated — дата создания анкеты
- NumberOfPictures — количество фотографий автомобиля
- PostalCode — почтовый индекс владельца анкеты (пользователя)
- LastSeen — дата последней активности пользователя

**Целевой признак**:
- Price — цена (евро)

А также **критерии**, которые важны заказчику:
- качество предсказания;
- время обучения модели;
- время предсказания модели.

```python
# Пропишем необходимые функции

def data_info(df):
    '''Информации о DF. Методы: describe, info, с просмотром памяти'''
    print(f'Размеры DF: {df.shape}', '\n')
    display(df.head(10), df.describe(), df.info(memory_usage='deep'))

def data_spread(column_list, df):
    '''Разброс данных. Отображение диаграммы размаха и гистограммы. Входные данные list колонок и df'''
    for column in column_list:
        df.boxplot(column=column)
        plt.title(f'Диаграмма размаха на столбце {column}')
        plt.show()
        df[column].plot(kind='hist', bins=100, grid=True)
        plt.title(f'Гистограмма на столбце {column}')
        plt.xlabel(column)
        plt.ylabel('Кол-во клиентов')
        plt.show()

def uniqueness(colomn_list, df):
    '''Просмотр уникальных значений листа'''
    for column in colomn_list:
        print(f'Количество уникальных значений {column}: {df[column].nunique()}')
        print(df[column].unique())
        print()
        
def value_index(df, column, arg):
    '''Функция, которая выводит список индексов по заданному аргументу'''
    return data.loc[data[column] == arg].index

def learning_time(model, features_train, target_train):
    '''Считает время обучения модели'''
    start = time.time()
    model.fit(features_train, target_train)
    end = time.time()
    learning_time = end - start
    return model, learning_time

def predict_time(model, features_test):
    '''Считает время предсказания модели'''
    start = time.time()
    predicts = model.predict(features_test)
    end = time.time()
    predict_time = end - start
    return predicts, predict_time

def alter_table(df, model, rmse, learning_time, predic_time):
    '''Добавляет в нашу таблицу с анализом моделей данные'''
    
    analytics.loc[model, 'RMSE'] = rmse
    analytics.loc[model, 'LearningTime'] = learning_time
    analytics.loc[model, 'PredictTime'] = predic_time
    
    return df
```

```python
# Удалим явные дубликаты

data = data.drop_duplicates().reset_index(drop=True)
print(f'Количество явных дубликатов = {data.duplicated().sum()}')
```

```python
# Заменим нулевые значения в столбце RegistrationMonth

# Посчитаем сколько нулевых данных можно поместить в каждый столбец, не ломая распределения
pr = []
for month in range(1, 13):
    prm = data.loc[data['RegistrationMonth'] == month].shape[0]/(data.loc[data['RegistrationMonth'] != 0].shape[0])
    ans = int(data.loc[data['RegistrationMonth'] == 0].shape[0] * prm)
    pr.append(ans)
    
# Проверка на корректность округления   
pr[11] = pr[11] - (sum(pr)-data.loc[data['RegistrationMonth'] == 0].shape[0])
assert sum(pr) == data.loc[data['RegistrationMonth'] == 0].shape[0], 'Перепроверить!'

# Узнаем индексы нулевых значений
index_RegistrationMonth = value_index(data, 'RegistrationMonth', 0)

# Поменяем нулевые значения на данные, в зависимости от их распределения
for month,i in enumerate(pr):
    index = np.random.choice(index_RegistrationMonth, size=i, replace=False)
    data.loc[index,'RegistrationMonth'] = (month+1)
    index_RegistrationMonth = index_RegistrationMonth[~np.isin(index_RegistrationMonth, index)]
    
# Выведем график сезонности регистрации автомобилей (продаж)
plt.figure(figsize=(15,15))
sns.distplot(data['RegistrationMonth'], bins=12)
plt.xlabel('Месяц')
plt.ylabel('Распределение')
plt.title('Сезонность регистрации АВТО')
plt.show()
```

Когда мы заполнили нулевые значения столбца RegistrationMonth от 1 до 12, исходя из распределения в выборке, нам удалось оставить ту сезонность продаж автомобилей, которую мы можем видеть выше на графике: спад активности приходит на зиму, а её пик - весна. ## Подготовка данных

```python
# Тк нам важно время обучения модели будем использовать технику порядкового кодирования (OE)

target = data['Price']
features = data.drop('Price', axis=1)
features_train, features_test_valid, target_train, target_test_valid = train_test_split(features, target, test_size=0.3, random_state=12345)
features_test, features_valid, target_test, target_valid = train_test_split(features_test_valid, target_test_valid, test_size=0.5, random_state=12345)


# Проверим на корректность
print(f'train {features_train.shape, target_train.shape}, valid {features_valid.shape, target_valid.shape}, test {features_test.shape, target_test.shape}')
```

```python
to_encode = ['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand', 'Repaired']
encoder = OrdinalEncoder()
encoder.fit(features_train[to_encode]) 
features_train[to_encode] = encoder.transform(features_train[to_encode])
features_test[to_encode] = encoder.transform(features_test[to_encode])
features_valid[to_encode] = encoder.transform(features_valid[to_encode])
display(features_train.head(3), features_test.head(3), features_valid.head(3))
```

```python
# Маштабируем признаки 

# Я решил не маштабировать RegistrationYear, Power , Kilometer и Repaired, тем самым выделив их и предав
# более сильное значение для модели, нежели остальные, также, чтобы выделить среди них самого важного напишем функцию после
numeric=['VehicleType', 'Gearbox', 'Model', 'RegistrationMonth', 'FuelType', 'Brand', 'PostalCode']

scaler = StandardScaler().fit(features_train[numeric])

features_train[numeric] = scaler.transform(features_train[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])

pd.options.mode.chained_assignment = None

display(features_train.describe(), features_test.describe(), features_valid.describe())
```

```python
# Напишем функцию, чтобы выделить наши признаки

# df = data.copy()

def scaller(df, column, arg):
    """Функция, которая работает по формуле StandardScaler, но умножаетзначения колонки на определённый аргумент
        предавая признаку значимость при обучении"""
    z = ((df[column] - df[column].mean()) / st.stdev(df[column])) * arg
    return z

# Расставим колонки по степени значимости
main_features = ['Repaired', 'Kilometer', 'RegistrationYear', 'Power']

# Маштабируем наши признаки
for counter, column in enumerate(main_features):
    features_train[column] = scaller(features_train, column, 10**(counter+1))
    features_test[column] = scaller(features_test, column, 10**(counter+1))
    features_valid[column] = scaller(features_valid, column, 10**(counter+1))
    
display(features_train.describe())
display(features_test.describe())
display(features_valid.describe())
```
